***APLICACIÓN DE PROBABILIDAD***

Se Realiza el lanzamiento de un dado EQUILIBRADO JUSTO, de manera reiterada con las mismas condiciones, y se anotan sus resultados= CARA O CRUZ.

Se le pide al usuario que defina la cantidad N de lanzamientos y se calcula su frecuencia relativa= 'PROBABILIDAD' de obtener cara o cruz. Se obtiene los siguientes resultados:

```python
import numpy as np
n=int(input('ingrese la cantidad de simulación o lanzamientos: '))

#========= FORMA 1 DE SIMULAR LANZAMIENTO DE UN DADO============
dado=[]
cara=0
cruz=0
for i in range(n):
  dado.append(np.random.rand(1))
  dado1=np.array(dado)
  if dado1[i]<0.5:
    print('salio cara')
    print(dado[i])
    cara +=1
  else:
    print('salio cruz')
    print(dado[i])
    cruz +=1
```

```python
#  mostrar la probailidad frecuencista de el lanzamiento de un dado
prom_cara=cara/n
prom_cruz=cruz/n
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CARA es:', prom_cara)
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CRUZ es:', prom_cruz)
```

```python
#========== FORMA COMPACTA DE SIMULAR USANDO NUMPY===========
m=int(input('ingrese la cantidad de simulación o lanzamientos: '))
lanzamientos = np.random.randint(0, 2, m)

# Calcular las probabilidades

probabilidad_cara = np.sum(lanzamientos ==0 ) / len(lanzamientos)
probabilidad_sello = np.sum(lanzamientos== 1) / len(lanzamientos)

print(f"Probabilidad de cara: {probabilidad_cara:.2f}")
print(f"Probabilidad de sello: {probabilidad_sello:.2f}")
```

```python
import matplotlib.pyplot as plt
x=np.array(range(2))
y=np.array([probabilidad_cara, probabilidad_sello])

plt.bar(x,y)
plt.xlabel('Opción')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma del lanzamiento de una moneda equilibrada')
plt.show()
```

##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt


probabilidad=1/6
np.random.seed(42)
num_lanzamientos=int(input('ingrese la cantidad de simulación o lanzamientos: '))
lanzamientos=np.random.choice(np.arange(1,7),size=num_lanzamientos, p=[probabilidad]*6)
print()
print(lanzamientos)
media=np.mean(lanzamientos)
mediana=np.median(lanzamientos)
desviacion_standard=np.std(lanzamientos)
varianza=np.var(lanzamientos)
percentiles=np.percentile(lanzamientos,[25,50,75])

valores_individuales=[1,2,3,4,5,6]
print()

resultados=[]
for i in valores_individuales:
  count=np.sum(lanzamientos == i)
  probab=round(count/num_lanzamientos,2)
  resultados.append(probab)
  print('La probabilidad de obtener ', i,  'es de ', probab)

print('los resultados de los lanzamientos son: ',resultados)
x=valores_individuales
y=resultados
plt.bar(x,y,color='lightsalmon')
for i, probabilidad in enumerate(resultados):
  plt.text(valores_individuales[i], probabilidad + 0.02, f'{probabilidad:.2f}', ha='center')
plt.xlabel('Número')
plt.ylabel('Probabilidad')
plt.title('Histograma de Probabilidades del dado equilibrado')
plt.axvline(media,color='red',label='media')
plt.xticks(valores_individuales)
plt.legend()
plt.show()

```

## EJERCICIO 3

 Realizar la simulación del juego {piedra,papel , tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probailidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidad. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
import numpy as np
import matplotlib.pyplot as plt

opciones=['piedra','papel','tijera']    #el juego tiene 3 opciones
N=int(input('ingrese la cantidad de jugadas: '))   # el usuario puede introducir el número de jugadas

probabilidad=1/3   #la probabilidad es equilibrada
np.random.seed(42)
lanzamientos=np.random.choice(opciones,size=N)
print(lanzamientos)   #se imprimen las opciones que hubo en las jugadas

resultados=[]    # aquí se almacenarán los resultados
for i in opciones:   #se hace un ciclo for para estudiar la probabilidad de cada opción
  count=np.sum(lanzamientos == i)   #cuenta la cantidad en que salió cada opción
  probab=round(count/N,2)    # la probabilidad es el resultado anterior, dividido por l a cantidad de jugadas (y se usan 2 decimales)
  resultados.append(probab)   # se van agregando los resultados
  print('La probabilidad de obtener ', i,  'es de ', probab)


x=opciones
y=resultados

plt.bar(x,y,color='lightsalmon')
for i, probabilidad in enumerate(resultados):
  plt.text(opciones[i], probabilidad + 0.02, f'{probabilidad:.2f}', ha='center')
plt.xlabel('Número')
plt.ylabel('Probabilidad')
plt.title('Histograma de Probabilidades')
plt.xticks(opciones)
plt.legend()
plt.show()

print('Se hizo la prueba jugando diferente cantidad de veces. A mayor cantidad de jugadas N, la probabilidad tiende a parecerse; mientras que si hay pocas jugadas, la probabilidad es muy dispareja')

```
