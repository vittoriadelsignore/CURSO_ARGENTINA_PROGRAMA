Solicitar al usuario que ingrese un número, si es negativo o nulo , solicitar reiteradamente hasta que ingrese un numero positivo.

y comparamos si es positivo. Si no lo es hay que pedir otro dato, pero puede ser nulo, entonces habría que hacer todo de nuevo. Por eso llamo en el  **else** a la funcion  **positivo**, para que inicie todo de nuevo.
recuerden que para ejecutar una funcion hay que llamarla con su nombre y parentesis. Ejemplo: positivo()

```python
def numero_positivo():
  while True:
    a=int(input('ingrese un número: '))
    if a<= 0:
      print('no es mayor que 0')
    else:
      return a

print('el numero del usuario es: ', numero_positivo())
```

**EJERCICIO 8 LISTA CON CICLO FOR**

Solicitar a 10 usuario que ingresen los siguientes datos en el formato lista.

['NOMBRE','APELLIDO', 'DNI','GASTOS MENSUALES DE HOGAR']

Ingresar el final de la lista los siguientes usuarios:

['ESTEBAN','LOPEZ', '45675400','190800.80']

['NOELIA','ALEJA', '46906875','85900.50']

Mostrar el vector.
Eliminar el primer usuario de la lista.

para probar como funciona cambien en el for el 10 por 2 y ejecuten ingrese los datos que le pidan y al terminar todo vuelvan a poner 10 en el for

```python
USUARIOS = []
for i in range(2):
    NOMBRE = input("Ingresa el nombre del usuario: ")
    APELLIDO = input("Ingresa el apellido del usuario: ")
    DNI = input("Ingresa el DNI del usuario: ")
    GASTOS_MENSUALES = float(input("Ingresa los gastos mensuales de hogar del usuario: "))
    USUARIO = [NOMBRE, APELLIDO, DNI, GASTOS_MENSUALES]
    USUARIOS.append(USUARIO)

print("El vector de usuarios es:")
for USUARIO in USUARIOS:
    print(USUARIO)


EL=['ESTEBAN','LOPEZ', '45675400','190800.80']

NA=['NOELIA','ALEJA', '46906875','85900.50']

USUARIOS.append(EL)
USUARIOS.append(NA)

print("El vector de usuarios es:")
for USUARIO in USUARIOS:
    print(USUARIO)


USUARIOS.pop(0)

print("EL VECTOR FINAL ES:")
for USUARIO in USUARIOS:
    print(USUARIO)
```

**EJERCICIO 9 LISTA CON CICLO FOR**

Buscar en la web los datos pablacionales de todas las provicias de Argentina , generando una lista con los siguiente información.

['PROVINCIA','CANTIDAD DE HABITANTES', 'CONSUMO ELECTRICO PER CAPITA','SUPERFICIE EN M^2']



Mostrar TODOS LOS DATOS DEL LISTADO.



```python
import requests

def obtener_datos_poblacionales():
    url = "URL_DE_LA_API_O_SITIO_WEB"
    response = requests.get(url)

    if response.status_code == 200:
        datos_provincias = response.json()

        lista_datos = []
        for provincia in datos_provincias:
            nombre_provincia = provincia['provincia']
            cantidad_habitantes = provincia['habitantes']
            consumo_electrico_per_capita = provincia['consumo_electrico']
            superficie_m2 = provincia['superficie']

            lista_datos.append([nombre_provincia, cantidad_habitantes, consumo_electrico_per_capita, superficie_m2])

        return lista_datos
    else:
        print("Error al obtener los datos poblacionales.")
        return []

# Llamamos a la función para obtener los datos poblacionales
datos_poblacionales = obtener_datos_poblacionales()

# Mostramos TODOS LOS DATOS DEL LISTADO
for provincia in datos_poblacionales:
    print(provincia)
```
