####Ejercicio 2: Descripción del dataset

El dataset tiene información de clientes que han comprado seguros, separando la información de acuerdo al tipo de seguro que compró: salud, bicicleta, cajero, mascota o vida. Indica un valor de identificación del cliente, la fecha en que compró el producto, la fecha en que se generó y se le entregó el contrato con la póliza, su género, si es femenino o masculino, el grupo etario al que pertenece, el plan que decidió comprar, la provincia donde vive, la cantidad de primas que ha pagado, el monto que cuesta cada plan, el socio a quien se lo compró. En este caso son 2 socios que componen la empresa que vende los seguros al consumidor final.Y también está el ingreso total que ha generado cada cliente, que es un aproximado del monto mensual, y la cantidad de meses de antiguedad (aproximado porque hay comisiones pagadas a empleados).

####Ejercicio 3: Definición de cada variable

El dataset corresponde a las ventas que tiene una empresa de seguros.  En el se muestran:

*  FECHA_VENTA: formato fecha, información en que se vendió el seguro.
*  SOCIO_ID: variable numérica, discreta. Cada número corresponde al socio al que se realiza la venta.  Es una empresa conformada por 2 socios principales, y ellos a su vez le venden al consumidor final.
*  PRODUCTO: variable de texto, categórica nominal, muestra el producto vendido.
*  CANAL: variable de texto, categórica nominal, muestra el canal donde se realizó la venta (tienda, ecommerce, telemarketing).
*  CANT_PRIM: variable numérica, discreta. Muestra el conteo de la cantidad de primas que ha pagado el cliente.
*  GENERO: variable de texto, indica si es femenino o masculino.
*  GRUPO ETARIO: variable de texto, que muestra rangos de edades del cliente que compra el seguro.
*  VIGENCIA_DESDE: formato fecha, información en que se generó el documento con la póliza y se entregó al cliente.
*  Antiguedad: variable numérica, discreta. Muestra la cantidad de meses de vigencia del cliente.
*  PLAN: variable de texto, ordinal debido a que cada plan tiene un precio diferente, indica qué plan compró el cliente.
*  PROVINCIA: variable de texto, nominal. Indica la provincia donde se compró el seguro.
*  IDCLIENTE: variable numerica, discreta. Identificador del cliente.
*  ESTADO_VIGENCIA: variable de texto, nominal. Indica si el cliente aún tiene la póliza o se dio de baja.
*  MONTO: variable numérica, ordinal. Indica el precio que paga mensualmente el cliente. Depende del producto y plan comprado.
*  TOTAL INGRESO: variable numerica, discreta. Indica el ingreso total que ha dejado el cliente a la empresa, ya que cada periodo paga una mensualidad por el disfrute del producto o seguro.

```python
#Para evaluar valores atípicos, primero hacemos un boxplot de las principales variables:
numericas=['CANT_PRIM','ANTIGUEDAD','MONTO','TOTAL INGRESO']
for i in numericas:
  fig, axs = plt.subplots()
  sns.boxplot(data=df[i])
  plt.title(i)
```

El "TOTAL INGRESO" es la variable que tiene outliers.

```python
# Calcular el rango intercuartílico (IQR) para la columna 'Age'
Q1 = df['TOTAL INGRESO'].quantile(0.25)
Q3 = df['TOTAL INGRESO'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para detectar valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos en la columna 'Age'
outliers = df[(df['TOTAL INGRESO'] < lower_bound) | (df['TOTAL INGRESO'] > upper_bound)]

# Imprimir valores atípicos
print("Valores Atípicos:")
print(outliers)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df_pca=df[['CANT_PRIM', 'N_VENTAS','ANTIGUEDAD','MONTO','TOTAL INGRESO']]
df_pca.head()
#se eliminan variables que no aportan al análisis
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(df_pca)
```

```python
pca=PCA()
pca.fit(data_scaled)
```

```python
explained_variance=pca.explained_variance_ratio_
cumulative_explained_variance=np.cumsum(explained_variance)
```

```python
explained_variance.cumsum()
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,len(cumulative_explained_variance)+1),cumulative_explained_variance,marker="o")
plt.xlabel("Numero Componentes Principales")
plt.ylabel("Varianza explicada acumulativa")
plt.title("Varianza explicada acumulativa")
plt.show()
```

```python
#determiar el numero de componentes principales a retener
numero_decomponentes=3
```

```python
pca=PCA(n_components=numero_decomponentes)
data_pca=pca.fit_transform(data_scaled)
```

```python
componentes=pca.components_
```

```python
plt.figure(figsize=(10,6))
plt.bar(range(len(componentes[0])),componentes[0])
plt.xticks(range(len(componentes[0])),df_pca,rotation=90)
plt.xlabel("caracteristicas")
plt.ylabel("importancia")
plt.title("la importancia de caracteristicas en el primer componente principal")
plt.show()
```
